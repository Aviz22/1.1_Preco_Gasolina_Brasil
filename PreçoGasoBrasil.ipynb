#####4. As colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converter ambas colunas para Timestamp/Datetime

```python
df["DATA INICIAL"] = pd.to_datetime(df["DATA INICIAL"])
df["DATA FINAL"] = pd.to_datetime(df["DATA FINAL"])
type(df["DATA FINAL"].iloc[2])
```

#####5. Criar uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência

#####12. Obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro

```python
# pegar a ultima informação apresentada no ano para cada estado e depois filtrar o Rio de Janeiro
# criar uma coluna "MES"
# Criar uma variavel DFRIO (QUE VAI SER APENAS UM FILTRO NO ESTADO DO RIO)
# Fazer um agrupamento no ANO-MES e pegar apenas a ultima informação do estado do Rio. Por que? tem mais de uma informação no mesmo mes
# Fazer um filtro no mês 12
df2["MES"] = df2["DATA FINAL"].apply(lambda x: x.month)
df_rio = df2[df2["ESTADO"] == "RIO DE JANEIRO"]
```

```python
df_month_rio = df_rio.groupby("ANO-MES") [["PREÇO MÉDIO REVENDA", "MES"]].last()
```

```python
(df_month_rio[df_month_rio["MES"] == 12] / df_month_rio[df_month_rio["MES"] == 12].shift(1) -1) * 100
```

#####Desafio: Criar uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresentando ao lado os estados na qual os maiores e menores preços foram registrados

```python
# Agrupando o max e min do preço médio de revenda por ANO-MES
df_max = df2.groupby("ANO-MES").max() ["PREÇO MÉDIO REVENDA"]
df_min = df2.groupby("ANO-MES").min()["PREÇO MÉDIO REVENDA"]

idx_max = df2.groupby("ANO-MES") ["PREÇO MÉDIO REVENDA"].idxmax()
idx_min = df2.groupby("ANO-MES") ["PREÇO MÉDIO REVENDA"].idxmin()
```

```python
df_diff = pd.DataFrame()
```

```python
df_diff
```

```python
df_diff["abs_diff"] = df_max - df_min
df_diff["percent_diff"] = (df_max - df_min) / df_min * 100 #colocando no formato percent
df_diff["max"] = df_max
df_diff["min"] = df_min
```

```python
df_diff
```

```python
df2.loc[idx_max, :] ["ESTADO"].values # FAZER ISSO PORQUE O INDICE NAO ESTÁ IGUAL, POREM OS VALORES SIM
```

```python
df_diff["ESTADO_MAX"] = df2.loc[idx_max, :] ["ESTADO"].values
df_diff["ESTADO_MIN"] = df2.loc[idx_min, :] ["ESTADO"].values
```

```python
df_diff
```

```python
# ESTADO COM O MAIOR PREÇO

df_diff["ESTADO_MAX"].value_counts()
```
